### Additional Part: Histogram Test

```python
resid = 4
dihedral_name = 'C2prime-P'
x_container = np.rad2deg(plot_agent.d_reader.get_d_time_dihedral_by_resid(resid, dihedral_name)['dihedral']) + 180
x_container.shape
```

```python
prob_array, bin_array = np.histogram(x_container, bins=np.arange(0, 359.1, 1), density=True)
x_array = [(bin_array[idx]+bin_array[idx+1])/2  for idx in range(bin_array.shape[0]-1)]
sum(prob_array)
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(x_array, prob_array)
plt.show()
```

```python
yticklabels = np.arange(0, 359.1, 30)
yticklabels = [f'{label:.0f}' for label in yticklabels]
n_yticklabels = len(yticklabels)
interval = (top - bottom) / n_yticklabels
yticks = np.arange(bottom, top, interval)
```

### Test: $\epsilon-\zeta$

```python
prob_array = plot_agent.epsilon_minus_zeta_df['15']
xarray = np.arange(-159.5, 199, 1)
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(xarray, prob_array)
ax.axvline(20, color="red", alpha=0.2)
ax.set_ylabel('Probability')
ax.set_xlabel('$\epsilon-\zeta$')
plt.show()
```
